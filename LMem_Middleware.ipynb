```python
# [2/4] Dictionary (9,076 entries, embedded)

_B64 = (
    ""
)

DICT_ENTRIES = json.loads(zlib.decompress(base64.b64decode(_B64)).decode("utf-8"))
print(f"Dictionary loaded: {len(DICT_ENTRIES)} entries")
```

```python
# [3/4] Middleware engine

_enc = tiktoken.get_encoding("cl100k_base")
_bpe_cache = {}

def _bpe_cost(s):
    if s not in _bpe_cache:
        _bpe_cache[s] = len(_enc.encode(s))
    return _bpe_cache[s]

def token_count(text):
    """Return BPE token count for text"""
    return len(_enc.encode(text))

def compress(code):
    """DP-optimal compression -> (compressed_text, used_entries)"""
    N = len(code)
    if N == 0:
        return code, []
    safe = [e for e in DICT_ENTRIES if e["symbol"] not in code]
    starts = defaultdict(list)
    for entry in safe:
        pat = entry["pattern"]; idx = 0
        while True:
            pos = code.find(pat, idx)
            if pos == -1: break
            starts[pos].append((pos + len(pat), entry)); idx = pos + 1
    if not starts:
        return code, []
    INF = float("inf"); dp = [INF]*(N+1); choice = [None]*(N+1); dp[N] = 0
    for i in range(N-1, -1, -1):
        for end, entry in starts.get(i, []):
            sc = _bpe_cost(entry["symbol"])
            if dp[end]+sc < dp[i]: dp[i]=dp[end]+sc; choice[i]=(end, entry)
        cc = _bpe_cost(code[i])
        if dp[i+1]+cc < dp[i]: dp[i]=dp[i+1]+cc; choice[i]=(i+1, None)
        for cl in range(2, min(11, N-i+1)):
            ch = code[i:i+cl]; chc = _bpe_cost(ch)
            r = dp[i+cl] if i+cl<=N else 0
            if r+chc < dp[i]: dp[i]=r+chc; choice[i]=(i+cl, None)
    used_set = set(); segments = []; pos = 0
    while pos < N:
        np_, entry = choice[pos]
        if entry: segments.append(entry["symbol"]); used_set.add(id(entry))
        else: segments.append(code[pos:np_])
        pos = np_
    compressed = "".join(segments)
    used = [e for e in safe if id(e) in used_set]
    if decompress(compressed, used) != code:
        # Greedy fallback
        current = code; used = []
        for entry in sorted(DICT_ENTRIES, key=lambda x: -len(x["pattern"])):
            pat, sym = entry["pattern"], entry["symbol"]
            if sym in code or sym in current: continue
            if pat in current: current = current.replace(pat, sym); used.append(entry)
        return current, used
    return compressed, used

def decompress(compressed, used_entries):
    """Restore compressed text to original (lossless)"""
    sym_map = {e["symbol"]: e["pattern"] for e in used_entries}
    if not sym_map: return compressed
    sorted_syms = sorted(sym_map.keys(), key=len, reverse=True)
    result = []; i = 0
    while i < len(compressed):
        matched = False
        for sym in sorted_syms:
            if compressed[i:i+len(sym)] == sym:
                result.append(sym_map[sym]); i += len(sym); matched = True; break
        if not matched:
            result.append(compressed[i]); i += 1
    return "".join(result)

def process(code, show=True):
    """Compress + restore + verify in one call"""
    orig = token_count(code)
    compressed, used = compress(code)
    comp = token_count(compressed)
    restored = decompress(compressed, used)
    ok = restored == code
    pct = (1 - comp/orig)*100 if orig > 0 else 0
    if show:
        print(f"Before:  {orig:>6} tokens  ({len(code)} chars)")
        print(f"After:   {comp:>6} tokens  ({len(compressed)} chars)")
        print(f"Saved:   {orig-comp:>6} tokens  ({pct:.1f}%)")
        print(f"Lossless: {'PASS' if ok else 'FAIL'}")
    return {"compressed": compressed, "used": used, "original_tokens": orig,
            "compressed_tokens": comp, "reduction_pct": round(pct,2), "lossless": ok}

print("Middleware ready. Use: process(your_code)")
```

```python
# [4/4] Paste your code here and run

MY_CODE = """\
from pathlib import Path
from typing import Optional, List, Dict, Any
import json
import logging


class DataProcessor:
    def __init__(self, config: dict):
        self.config = config
        self.base_dir = Path(config.get("base_dir", "."))
        self._results: dict[str, Any] = {}
        self._logger = logging.getLogger(__name__)

    def process(self, file_path: str) -> Optional[dict]:
        try:
            with open(file_path, "r") as f:
                data = json.load(f)
            self._results[file_path] = data
            self._logger.info(f"Processed: {file_path}")
            return data
        except Exception as e:
            self._logger.error(f"Error: {file_path}: {e}")
            return None

    def summary(self) -> None:
        total = len(self._results)
        print(f"Total processed: {total}")


def main():
    logging.basicConfig(level=logging.INFO)
    processor = DataProcessor({"base_dir": "/tmp/data"})
    processor.process("input.json")
    processor.summary()


if __name__ == "__main__":
    main()
"""

result = process(MY_CODE)

# result contains everything needed for restoration:
#   result["compressed"]  -> compressed text
#   result["used"]        -> dictionary entries used
```

```python
# --- Method 2: Save/load compression result as JSON ---

import json

# Save (after compressing)
save_data = {
    "compressed": result["compressed"],
    "used": result["used"],
}
with open("compressed_output.json", "w", encoding="utf-8") as f:
    json.dump(save_data, f, ensure_ascii=False, indent=2)
print("Saved to compressed_output.json")

# Load and restore
with open("compressed_output.json", "r", encoding="utf-8") as f:
    loaded = json.load(f)
restored = decompress(loaded["compressed"], loaded["used"])
print(f"\nRestored ({len(restored)} chars):")
print(restored[:300])
print(f"Lossless: {'PASS' if restored == MY_CODE else 'FAIL'}")
```

## API helper (optional)

Use these when integrating with LLM APIs (Claude, GPT, etc.).

```python
# Optional: API integration helpers

def compress_for_api(code):
    """Compress code before sending to LLM API"""
    return compress(code)

def decompress_from_api(compressed, used):
    """Decompress LLM response back to original"""
    return decompress(compressed, used)

def make_dict_prompt(used):
    """Generate a system prompt so the LLM can read compressed code"""
    lines = ["The following code uses symbol compression. Each symbol maps to a pattern:\\n"]
    for e in used:
        lines.append(f"{e['symbol']} = {repr(e['pattern'])}")
    return "\\n".join(lines)

# --- Demo ---
demo_code = """def hello(name: str) -> str:\n    return f\"Hello, {name}!\"\n"""
compressed, used = compress_for_api(demo_code)
print("=== Compressed (send this to API) ===")
print(compressed)
print(f"\nTokens: {token_count(demo_code)} -> {token_count(compressed)}")
print(f"\n=== Dict prompt (add to system prompt) ===")
print(make_dict_prompt(used)[:500])
print(f"\n=== Restored ===")
restored = decompress_from_api(compressed, used)
print(restored)
print(f"Lossless: {'PASS' if restored == demo_code else 'FAIL'}")
```
